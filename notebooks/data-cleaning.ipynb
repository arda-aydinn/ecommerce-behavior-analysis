```python
import pandas as pd
import numpy as np
```

```python
# ---GET TO KNOW THE DATA--- (Step 1 And Step 2)
df = pd.read_csv('dirty_synthetic_ecommerce_tr_1700.csv')

print("---Data Info Here---")
df.info()

print("\n---First 5 Rows---")
display(df.head(5))

print("\n---Last 5 Rows---")
display(df.tail())
```

```python
# ---IF THIS DATA TIDY?--- (STEP 3)

#To check we have actual column names (variable names).
print("Current Column Names:")
print(list(df.columns))

#to see the structre of first 3 rows
display(df.head(3))
```

```python
# --- FIXING COLUMN NAMES --- (Step 4)

#first we see the original column names in order to check whether we did our job well or not at the end.
print("Original Column Names:")
print(list(df.columns))

#removing whitespaces here
df.columns = df.columns.str.strip()

#standardizing all variable names here
df.columns = df.columns.str.lower()

#replacing any spaces with underscores here
df.columns = df.columns.str.replace(" ", "_")

df = df.rename(columns={
    "membershiptier": "membership_tier",
    "basketvaluetry": "basket_value_try",
    "productcategory": "product_category",
    "satisfactionscore": "satisfaction_score",
    })

#the result must be shown here
print("\nCorrected And Fixed Column Names: ")
print(list(df.columns))
```

```python
# ---REMOVING UNNECESSARY COLUMNS--- (Step 5)

#as we will not take orderid column into consideration (mostly) we can drop it.
df = df.drop(columns= ["orderid"], errors="ignore") #to prevent any erros in jupiter notebook

print("\nCurrent Columns: ")
print(list(df.columns))
```

```python
# --- REMOVING DUPLICATES ---  (Step 6)

#we can verify at the end if we count our duplicated results first. also we need to check whether we have duplicates or not.
nDuplicates = df.duplicated().sum()
print(f"Number of duplicates: {nDuplicates}")

#we do not have duplicates yet; however, we did not standardize our table.
df = df.drop_duplicates()

#verifying the result here
print(f"All duplicates are deleted. New row count: {len(df)}")
```

```python
# --- REMOVING  UNNECESSARY CHARACTERS IN VALUES --- (Step 7)

# we will go by column by column. we can check if there are any unnecessary strings. then remove or replace them.
display(df.head(15)) # to see our data and detect the unnecesary characters.

# we do not need '%' sign here, so we remove it.
df['discount_rate'] = df['discount_rate'].astype(str).str.replace('%', '', regex=False) #regex is false here because we do not look for repeated values

# there are many * signs in our data. we can remove them bu using loops.
text_columns = df.select_dtypes(include=['object']).columns
for col in text_columns:
    if df[col].astype(str).str.contains('*', regex=False).any(): #any column that contains * sign in them
        print(f"Cleaning '*' from column: {col}") # to see which colum had * sign
        df[col] = df[col].astype(str).str.replace('*', '', regex=False)

display(df.head(15))
```

```python
# --- REMOVING WHITE SPACES --- (Step 8)

#selecting all string columns
space_columns = df.select_dtypes(include=["object"]).columns

#to see our data first
display(df[space_columns].head(5))

#using loop to remove all white spaces
for col in space_columns:
    df[col] = df[col].str.strip()

print("All white spaces are removed.")

#to verify our result
display(df[space_columns].head(5))
```

```python
# --- CONVERTING ALL TEXTS TO LOWERCASE --- (Step 9)

#selecting our string columns
nuCols = df.select_dtypes(include=["object"]).columns

#see our data first
display(df[nuCols].head(5))

#using for loop to change them
for col in nuCols:
    df[col] = df[col].str.lower()

#verifying the result
display(df[nuCols].head(5))
```

```python
# --- FIXING THE TYPOS --- (Step 10)

#defining all the values on our string columns and using loop to determine uniqueness
defCols = df.select_dtypes(include=["object"]).columns.drop('discount_rate')
for col in defCols:
    print(f"\n--- {col} ---")
    print(df[col].unique())

#we set a dictionary and replace the values
typo_corrections= {
    "f3mal3": "female",
    "mal3": "male",

    "m3rsin": "mersin",
    "gaziant3p": "gaziantep",
    "kays3ri": "kayseri",
    "eskiş3hir":"eskişehir",

    "silv3r": "silver",
    "gu3st": "guest",
    "bronz3": "bronze",

    "mobil3": "mobile",
    "d3sktop": "desktop",
    "tabl3t": "tablet",

    "paid s3arch": "paid search",
    "r3f3rral": "referral",

    "d3bit card": "debit card",
    "cr3dit card": "credit card",
    "bank transf3r": "bank transf3r",
    "digital wall3t":"digital wallet",
    "cash on d3liv3ry": "cash on delivery",
    "bank transf3r": "bank transfer",

    "books & station3ry": "books & stationery",
    "groc3ri3s": "groceries",
    "hom3 & kitch3n": "home & kitchen",
    "b3auty": "beauty",
    "el3ctronics": "electronics"
}
df = df.replace(typo_corrections)

#we check whether we achieve our aim or not using .unique() function
print("\nNew values: ")
for col in defCols:
    print(f"\n--- {col} ---")
    print(df[col].unique()) #now we have unique values as we can see

for col in defCols: 
    print()
    print(df[col].value_counts()) #we check the value counts here




```

```python
# --- COMBINING A DATE COLUMN  --- (Step 11)
#since we do not have month and day values (we only have year), we cannot create a date column.
```

```python
# --- FIXING DATA TYPES  --- (Step 12)

#checking the data types before the change
print(df.dtypes)

#changing the data types accordingly
df["year"] = df["year"].astype("Int64")
df["gender"] = df["gender"].astype(object)
df["age"] = df["age"].astype("Int64")
df["city"] = df["city"].astype(object)
df["membership_tier"] = df["membership_tier"].astype(object)
df["device_type"] = df["device_type"].astype(object)
df["traffic_source"] = df["traffic_source"].astype(object)
df["payment_method"] = df["payment_method"].astype(object)
df["product_category"] = df["product_category"].astype(object)
df["items_in_cart"] = df["items_in_cart"].astype("Int64")
df["basket_value_try"] = df["basket_value_try"].astype("float64")
df["discount_rate"] = df["discount_rate"].astype("float64")
df["delivery_days"] = df["delivery_days"].astype("float64")
df["satisfaction_score"] = df["satisfaction_score"].astype("float64")
df["repeat_purchase_90d"] = df["repeat_purchase_90d"].astype(bool)

#verifying the outputs
print()
print(df.dtypes)

```

```python
# --- DESCRIPTIVE STATISTICS --- (Step 13)
# using .describe() to access statistics of the data
stats = df.describe().T
print(stats)

#from the table we can say that there are missing values which will be handled in step 16. Other than NaN values, there are no unusual behaviors.
```

```python
# --- DETECTING OUTLIERS --- (Step 14)

#defining the columns we want to search outliers
cols = df.select_dtypes(include=["Int64","float64"]).columns.drop("year")
for col in cols:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1 

    outliers = (df[col] < q1 - 1.5*iqr) | (df[col] > q3 + 1.5*iqr) #defining outliers
    num_outliers = outliers.sum()

    if num_outliers > 0: #if we detect outliers
        mean_val = df[col].mean()

        if str(df[col].dtype) == "Int64": #if the type is int64 then we will also change the mean value to prevent errors
            mean_val = int(mean_val)
        else:
            mean_val = round(mean_val, 2) #we will round floats up to 2 decimals
        df.loc[outliers, col] = mean_val
        print(f"{col}: found {num_outliers} outliers and changed them with mean value ({mean_val}).")
    else:
        print(f"{col}: No outliers detected.")

        
```

```python
# --- CHECKING UNIFORMITY --- (Step 15)

#for categorical ones we can check if the values are unique
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    print(f"\nColumn: {col}")
    print(f"Unique Values ({len(df[col].unique())}):")
    print(df[col].unique()) 

#for numerical values we can check min and max values of each column
num_cols = df.select_dtypes(include=['number']).columns.drop('year', errors='ignore')

for col in num_cols:
    print(f"\nColumn: {col}")
    print(f"Min: {df[col].min()}  |  Max: {df[col].max()}") #if all values are within the domain of each column, then we achieve uniformity
```

```python
# --- DETECTING NAN VALUES AND REPLACING THEM (IMPUTATION) --- (Step 16)

# replacing all string/nan/NA values with standardized numpy NaN --> this was the most important step I think
df = df.replace(["NA", "na", "N/A", "", "nan"], np.nan)

nNaN = df.isnull().sum()
NaNRatio = (nNaN / len(df)) * 100

print("Missing Value Ratios (%):")
print(NaNRatio[NaNRatio > 0].round(2)) #show only columns with missing values

# imputation part is here
for col in df.columns:
    if df[col].isnull().sum() > 0: # dealing with only columns with NaN values
        
        # if the column is a string value, we fill with the mode value
        if df[col].dtype == "object" or df[col].dtype.name == "category":
            mode_val = df[col].mode()[0]
            df[col] = df[col].fillna(mode_val)
            print(f"{col:<20} -> Filled with MODE: '{mode_val}'")
            
        #if the column is a numerical one, we fill with the mean value
        else:
            mean_val = df[col].mean()
            if "int" in str(df[col].dtype).lower():
                fill_val = int(mean_val) #if it's an integer fill it with mean
            else:
                fill_val = round(mean_val, 2) #round to 2 decimals if it is a float
            
            df[col] = df[col].fillna(fill_val)
            print(f"{col:<20} -> Filled with mean: {fill_val}")

print("\n")
print("FINAL CHECK: Total Remaining NaN Values:", df.isnull().sum().sum())
```

```python
# Final Step
# We want to save our file as a new file.
df.to_csv("cleaned_ecommerce_data.csv", index=False)
```
