```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_ecommerce_data.csv")

#RQ: Which category has the highest average baske value, and how does it differ among genders?

sns.set_style("whitegrid") #to make it more readable
plt.figure(figsize=(12, 6)) #setting the figure size


order_list = df.groupby("product_category")["basket_value_try"].mean().sort_values(ascending=False).index #defining an order list

graph = sns.barplot(
    data=df, 
    x="product_category", #categorical variable on x axis
    y="basket_value_try",  #numerical variable on y axis
    hue="gender",  #third variable represents colors
    palette="viridis", 
    errorbar=None, #to make it more readable
    order=order_list
)

for container in graph.containers: 
    graph.bar_label(container, fmt="%.1f", fontsize=10) #to have labels on each bar

plt.title("Average Basket Value by Product Category and Gender", fontsize=15, fontweight="bold") #adding title
plt.xlabel("Product Category", fontsize=12) #setting label for x axis
plt.xticks(rotation=45) #rotating to prevent confusion

plt.ylabel("Average Basket Value (TRY)", fontsize=12) #setting label for y axis
plt.legend(title="Gender", loc="upper right") #setting a legend

plt.savefig("multivariate_barplot_category_gender.png", bbox_inches="tight") #saving the figure
plt.show() #showing the plot
```
